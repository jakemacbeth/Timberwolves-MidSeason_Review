```python
import sys
from pathlib import Path
import pandas as pd
from sqlalchemy import text
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = "https://raw.githubusercontent.com/jakemacbeth/Timberwolves-MidSeason_Review/main/data/"

wolves_df = pd.read_csv(base + "teambox_bucket_teams.csv")
opp_df = pd.read_csv(base + "opponnent_boxscores_bucketed.csv")
teambox = pd.read_csv(base + "teambox_pergame.csv")
```

```python
wolves_df.columns.tolist()
```

```python
wolves_df = wolves_df.drop(columns=[
 'game_id-2',
 'season',
 'is_home',
 'minutes',
 'last_updated_at']
)
```

```python
wolves_df.columns.tolist()
```

```python
opp_df.columns.tolist()
```

```python
opp_df = opp_df.drop(columns=[
 'opponent_team_name',
 'season',
 'opponent_team_id',
 'last_updated_at']
)
```

```python
opp_df.columns.tolist()
```

```python
stat_cols = ['pts',
 'fgm',
 'fga',
 'fg3m',
 'fg3a',
 'ftm',
 'fta',
 'oreb',
 'dreb',
 'reb',
 'ast',
 'stl',
 'blk',
 'tov',
 'pf',
 'off_rating',
 'def_rating',
 'net_rating',
 'pace',
 'ts_pct']

bucket_means = (
    opp_df
    .groupby(['opp_bucket', 'wolves_win'])[stat_cols]
    .mean()
    .reset_index()
    .round(1)
)
bucket_means
```

```python
diagnostic_df = bucket_means.drop(columns=['pace','ts_pct', 'net_rating', 'oreb', 'pf'])
diagnostic_df
```

```python
results = diagnostic_df.drop(columns=['dreb', 'reb', 'blk', 'tov', 'stl', 'ast', 'off_rating', 'def_rating', 'pts'])
results.head()
```

```python
wolves_id = 1610612750

# filter first
teambox_filtered = teambox[
    (teambox['opponent_team_id'] != wolves_id) &
    (teambox['opponent_team_id'].isin(opp_df['team_id']))
]

# bring bucket and win labels
opp_keys = opp_df[['team_id','wolves_win','opp_bucket']].drop_duplicates()

teambox_filtered = teambox_filtered.merge(opp_keys, on='team_id', how='left')

stats = ['fgm','fga','fg3m','fg3a','ftm','fta']

team_avg = (
    teambox_filtered
    .groupby(['opp_bucket','wolves_win'])[stats]
    .mean()
    .round(1)
    .reset_index()
)
team_avg
```

```python
results_fmt = results.rename(columns={
    'opp_bucket': 'Opponent Bucket',
    'wolves_win': 'Wolves Win',
    'fgm': 'FGM',
    'fga': 'FGA',
    'fg3m': 'FG3M',
    'fg3a': 'FG3A',
    'ftm': 'FTM',
    'fta': 'FTA'
})
results_styled = (results_fmt
 .style
 .format(precision=2)
 .hide(axis='index')
 .set_table_styles([
     {'selector': 'thead th',
      'props': [('background-color', '#1f4e79'),
                ('color', 'white'),
                ('font-weight', 'bold'),
                ('text-align', 'center')]},
     {'selector': 'tbody td',
      'props': [('text-align', 'center')]}
 ])
 .set_properties(**{'border': '1px solid #ddd'})
 .set_table_attributes('style="border-collapse:collapse;width:100%;"')
 .apply(lambda _: ['background-color: #f6f8fa' if i % 2 else '' for i in range(len(results_fmt))],
        axis=0)
)
stats = ['FGM','FGA','FG3M','FG3A','FTM','FTA']

results_styled.background_gradient(
    subset=stats,
    cmap='Blues'
).format({col:"{:.1f}" for col in stats}).set_caption(
    "Opponent Shooting Averages vs Wolves")
```
